```python
[x for x in wn.synsets('pasdfas')]
```

```python
def getAttributesForSynset(synset):
    return({
        'name': synset.name(),
        'pos': synset.pos(),
        'definition': synset.definition(),
        'lexname': synset.lexname(),
        'examples': ';'.join(synset.examples())        
    })
    


def getAttributesForWord(word):
    wn_synsets = wn.synsets(word)
    if len(wn_synsets) > 0:
        synsets = [getAttributesForSynset(x) for x in wn_synsets]
        df = pd.DataFrame(synsets)
        df['sense_index'] = range(df.shape[0])
        df['word'] = word
        df = df.sort_values(by=['sense_index'])
        return(df[['word','sense_index','name','lexname','pos','definition','examples']])
    else:
        return(None)
```
